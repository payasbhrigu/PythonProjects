```python
import requests
from bs4 import BeautifulSoup

import pandas as pd

city = input("Enter Name of City:").lower()

url = "https://www.makaan.com/"+city+"-residential-property/buy-property-in-"+city+"-city?"

Array= []

#for i in range(0,5):
while True:
    r=requests.get(url)

    c=r.content

    soup = BeautifulSoup(c,"html.parser")

    All = soup.find_all("div",{"class":"title-line"})

    for item in All:
        propUrl = item.find('a').get('href')
        r=requests.get(propUrl)
        c = r.content
        soup2 = BeautifulSoup(c,"html.parser")
        dic = {}
        try:
            dic["Type"] = soup2.find('span',{'class':'type'}).text
        except Exception:
            dic["Type"] = None
        try:
            dic["Builder(s)"] = soup2.find('a',{'class':'loclink'}).find('span').text
        except Exception:
            dic["Builder(s)"] = None
        try:
            dic['Price'] = soup2.find('span',{'class':'val'}).text.split()[0]+' '+soup2.find('span',{'class':'unit'}).text
        except Exception:
            dic['Price'] = None
        try:
            dic['Price(per sqft)'] = soup2.find('div',{'class':'rate-wrap'}).text.split()[0]+' '+'sq ft'
        except Exception:
            dic['Price(per sqft)'] = None
        try:
            dic['Status'] = soup2.find('td',{'id':'Status'}).text
        except Exception:
            dic['Status'] = None
        try:
            dic['Bathrooms'] = soup2.find('td',{'id':'Bathrooms'}).text
        except Exception:
            dic['Bathrooms'] = None
        try:
            dic['Floor No'] = soup2.find('td',{'id':'Floor'}).text.split()[0]
        except Exception:
            dic['Floor No'] = None
        try:
            dic['Total Floors'] = soup2.find('td',{'id':'Floor'}).text.split()[-1]
        except Exception:
            dic['Total Floors'] = None
        try:
            dic['Balconies'] = soup2.find('td',{'id':'Balconies'}).text
        except Exception:
            dic['Balconies'] = None
        try:
            dic['Additional Room(s)'] = soup2.find('td',{'id':'Additional Rooms'}).text
        except Exception:
            dic['Additional Room(s)'] = None
        try:
            dic['Booking Amount'] = soup2.find('td',{'id':'Booking Amount'}).text.split()[0]
        except Exception:
            dic['Booking Amount'] = None
        Array.append(dic)
        #print(dic)
    url = soup.find('div',{'class':'pagination'}).findAll('a')[-1].get('href')
    if url is None:
        break


```

```python
df = pd.DataFrame(Array)
df.head()
#df.to_csv(city+'.csv')
```
