```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import imutils
```

```python
image = cv2.imread("opencv-tutorial/jp.png")
```

```python
plt.imshow(image)
```

```python
(h,w,d) = image.shape
h,w,d
```

```python
cv2.imshow("Jurassic Park",image)
cv2.waitKey(0)
```

```python
roi = image[:175,300:425]
cv2.imshow("ROI",roi)
cv2.waitKey(0)
```

```python
resize_img = cv2.resize(image,(200,200))
cv2.imshow("Resized",resize_img)
cv2.waitKey(0)
```

```python
center = (w//2,h//2)
M = cv2.getRotationMatrix2D(center,-45,1.0)
rotated = cv2.warpAffine(image, M, (w, h))
cv2.imshow("Rotate",rotated)
cv2.waitKey(0)
```

```python
blur = cv2.GaussianBlur(image,(11,11),0)
cv2.imshow("Blur",blur)
cv2.waitKey(0)
```

```python
output = image.copy()
cv2.rectangle(output,(320,60),(420,160),(200,170,25),2)
cv2.imshow("Rectangle",output)
cv2.waitKey(0)
```

```python
output = image.copy(0)
cv2.putText(output,"Jurassic Park",(10,25),cv2.FONT_HERSHEY_SCRIPT_SIMPLEX,1,(256,125,150),2)
cv2.imshow("Text",output)
cv2.waitKey(0)
```

```python
cap = cv2.VideoCapture(0)
ret,img = cap.read()
cap.release()
img.shape
```

```python
#tetris = cv2.imread("opencv-tutorial/tetris_blocks.png")
cv2.imshow("Tetris",img)
cv2.waitKey(0)
```

```python
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2.imshow("Gray",gray)
cv2.waitKey(0)
```

```python
edged = cv2.Canny(img,150,160)
cv2.imshow("Edged",edged)
cv2.waitKey(0)
```

```python
thresh = cv2.threshold(gray,100,255,cv2.THRESH_BINARY_INV)[1]
cv2.imshow("Threshold",thresh)
cv2.waitKey(0)
```

```python
cnts = cv2.findContours(thresh.copy(), cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
output = img.copy()
cnts = cnts[0] if imutils.is_cv2() else cnts[1] 
# loop over the contours
for c in cnts:
    # draw each contour on the output image with a 3px thick purple
    # outline, then display the output contours one at a time
    cv2.drawContours(output, [c], -1, (240, 0, 159), 3)
cv2.imshow("Contours", output)
cv2.waitKey(0)
```

```python
text = "I found {} objects in this image".format(len(cnts))
cv2.putText(output, text, (10, 25),  cv2.FONT_HERSHEY_SIMPLEX, 0.7,(240, 0, 159), 2)
cv2.imshow("Contours", output)
cv2.waitKey(0)
```
